# 3. What are the two main factors that influence the behaviour of a 'normal' probability distribution?

The two main factors that influence the behaviour of a 'normal' probability distribution are :

1) Mean(μ) - Central Tendency :

   Determine the center of the bell curve.

   It represent the average value of the data set.

   Shifts the curve right(for higher mean) or left(for lower mean) along x-axis.

2) Standard Deviation (σ) - Spread or Width :

   This control how spread out or narrow the distribution is.

   A small (σ) means the curve is tall and narrow (most values are close to the mean).

   A large (σ) means the curve is flat and wide (values are more spread out from the mean).

# 5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term a probability distribution may show random variation or unexpected results.

As the number of trials grows :

1) The obeserved outcomes start to closely match the expected probabilities.
2) The distribution becomes more stable and start resembling the theoretical probability distribution.
