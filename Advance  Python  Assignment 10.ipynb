# 1. What is the difference between __getattr__ and __getattribute__?

The differnce between __getattr__ and __getattribute__ is :

__getattr__ is only called as a fallback when an attributes Not found through the normal lookup.
    
This is safer and more common choice because it trigger only for missing attributes.
    
Useful for handling missing attributes.

__getattribute__ is called when need to intercept every single attribute access.
    
This is more powerful but use it carefully because it can cause infinite recursion.
    
Good for debugging, logging or overriding all access behavior.

# 2. What is the difference between properties and descriptors?

The difference between properties and descriptors is as follow :

1) Properties :

    A properties is a built-in descriptor in python.

    It defined with @property decorator.

    It used for simple, class-specific attribute control.
    
    It allows easy creation of read/write property using getter, setter and deleter methods on the class.

    It implicitly handles one attribute at a time within the class.

2) Descriptors :

    A descriptor is a general protocol for customizing attribute access.
   
    It is the base meachanism behind properties and other tools.
        
    It defined with __get__(),__set__(),__delete__() .
        
    It used for more advanced attribute control or reusable logic across multiple classes or attributes.
        
    It explicitly reusable and can manage many attributes with a single descriptor class.
